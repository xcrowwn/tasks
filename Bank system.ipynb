```python
class User:
    def __init__(self,name,ID,gander,BD,address,email,phone_number,password,user_name):
            self.name=name
            self.ID=ID
            self.gander=gander
            self.BD=BD
            self.address=address
            self.email=email
            self.phone_number=phone_number
    def peresent(self):
        return f" name :{self.name} \n ID : {self.ID} \n BD {self.BD} \n gander : {self.gander} \n email{ self.email} \n phone number :{self.phone_number}"  
```

```python
class Bank(User):
    def __init__(self,name,ID,gander,BD,address,email,phone_number,balnce,acount_number,password,user_name):
        self.acount_number=acount_number
        self.password=password
        self.user_name=user_name
        self.balnce=float(balnce)
        super().__init__(name,ID,gander,BD,address,email,phone_number,password,user_name)
    def deposit(self,amount):
        amount=float(amount)
        self.balnce=self.balnce+amount
    def withdraw(self,amount):
        amount=float(amount)
        if self.balnce >= amount:
            self.balnce=self.balnce-amount
    def balnce_amount(self):
        return self.balnce
    
    def peresent(self):
        return f" name :{self.name} \n id : {self.ID} \n BD {self.BD} \n gander : {self.gander} \n email : { self.email} \n phone_number : {self.phone_number} \n acount_number : {self.acount_number} \n password : {self.password} \n user_name: {self.user_name} \n balnce : {self.balnce} "    

    def PasswordChecker(self,x,z):
        if x !=self.password and z!=self.user_name:
            return False
        else:
            return True
                
```
